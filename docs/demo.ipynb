```python
from datavyz import graph_env
import numpy as np

def fig_demo(ge):
    fig, AX = ge.figure((2,4), hspace=0.1)
    for i in range(4):
        y = np.random.randn(100)
        ge.scatter(y, range(len(y)), ax=AX[0][i], axes_args=dict(spines='left', ylabel='obs. #'));
        ge.hist(y, ax=AX[1][i], axes_args=dict(xlabel='value'));
        ge.title(AX[0][i], 'Sample %i' % (i+1))
    return fig, AX
```

```python
gedn = graph_env('dark_notebook')
fig, AX = fig_demo(gedn)
```

```python
# now same figure with the "manuscript" setting
gem = graph_env('manuscript')
fig, AX = fig_demo(gem)
```
